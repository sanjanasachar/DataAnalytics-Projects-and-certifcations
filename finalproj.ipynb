```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Count Plot
relation: how many Stars are there, in the rows I am focusing on

```python
dataset = pd.read_excel('amazon(for finalproj).xlsx', skiprows=range(8, 2213))
sns.countplot(x=dataset["Stars"])
sns.set(rc={'figure.figsize':(11.7,8.27)})
```

# Scatter Plot 
Relation: Does the amount of stars affect the discount price?
answer is no because even with a high stars the product is still selling at a very low discounted price. 

```python
dataset=pd.read_excel('amazon(for finalproj).xlsx', skiprows=range(8,2213))
sns.relplot(data=dataset,x='Stars', y='Discounted_Price',height=8.67, aspect=11.7/8.67)
```

# Line graph 
Relation: does the name of the product affect how much the discount price is?
no it doesn't because there isn't a relation on the graph. It is just linear which is expected. It must be based on the demand on each product that will explain the discounted price.  

```python
dataset=pd.read_excel('amazon(for finalproj).xlsx' skiprows=range(8,2213))
ax=sns.relplot(kind='line', data=dataset, x='Product_name', y='Discounted_Price', height=20.67, aspect=20.67/8.67)
```
